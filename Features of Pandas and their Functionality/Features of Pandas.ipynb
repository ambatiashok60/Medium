```python
import datetime
import calendar
```

### Function which converts the date into day name

```python
def convert_date_to_day_name(date):
    year, month, day = date.split("-")
    dt_idx = datetime.date(int(year),int(month),int(day)).weekday()
    return calendar.day_name[dt_idx]
```

### Apply and Lambda Operators

use convert_date_to_day_name function on the df["Date"] column to generate name of the day on those respective days

```python
df["Day_Name"] = df.apply(lambda x:convert_date_to_day_name(x["Date"]),axis=1)
df.head()
```

#### Extract the number of Mondays that have better opening stock value than previous Friday

```python
monday_df = df.groupby("Day_Name").get_group("Monday")
monday_df.head()
```

```python
friday_df = df.groupby("Day_Name").get_group("Friday")
friday_df.head()
```

```python
def convert_into_datetime_format(date):
    year,month,day = date.split("-")
    return datetime.datetime(int(year),int(month),int(day))
```

```python
import numpy as np
```

```python
result = []
for monday in monday_df.values:
    for friday in friday_df.values:
        if((convert_into_datetime_format(monday[0]) - convert_into_datetime_format(friday[0])).days == 3):
            result.append(monday[1]-friday[4])
```

```python
result = np.array(result)
sum(result>=0)
```

#### Extract the number of records at the end of the week that have better closing stock value than the closing value at the start of the week

```python
result = []
for monday in monday_df.values:
    for friday in friday_df.values:
        if(((convert_into_datetime_format(friday[0]) - convert_into_datetime_format(monday[0])).days == 4)):
            result.append(friday[4] - monday[1])
result = np.array(result)
sum(result>=0)
```
